```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

url = 'https://bit.ly/2SfTDe5'
df = pd.read_csv(url)

data = pd.DataFrame(df)
X = data.iloc[:, [1,2,3,4,5,6,7]].values
Y = data.iloc[:, 8].values

target = pd.DataFrame(Y).to_numpy()

def normalisasi_linear(inputList):
    result = inputList / np.linalg.norm(inputList)
    return result 
tempDf = normalisasi_linear(X)
df = tempDf

msk = np.random.rand(len(df)) < 0.75
data_train = df[msk]
data_test = df[~msk]
target_train = target[msk]
target_test = target[~msk]

print("Data training:", len(data_train))
print("Data test:", len(data_test))

lr = 0.01                                                 # alpha
maxep = 600000
nparam = len(data_train[0])
nhid = 10                                                  # jmlh hidden layer
noutput = 1                                               # Jml neuron di output layer

W1 = 2 * np.random.rand(nparam, nhid) - 1
W2 = 2 * np.random.rand(nhid, noutput) - 1
# print(W1)
# print(W2)

A1 = 1 / (1 + np.exp (-(np.dot(data_train, W1))))
A2 = 1 / (1 + np.exp(-(np.dot(A1, W2))))
print("cek tebakan awal sebelum training")
#print(A2)

print("Start training JST dengan jumlah hiden neuron = %d, jml output neuron = %d"%(nhid, noutput))
for i in range(maxep):
  #propagasi maju
    A1 = 1 / (1 + np.exp (-(np.dot(data_train, W1))))
    A2 = 1 / (1 + np.exp(-(np.dot(A1, W2))))
  #hitung error
    error = target_train - A2
    #print(error)

  #Propagasi mundur (update bobot)
    D2 = error * (A2 * (1 - A2))
    D1 = D2.dot(W2.T) * (A1 * (1 - A1))
    
    W2 += A1.T.dot(D2) * lr
    W1 += data_train.T.dot(D1) * lr

print("Selesai training")

#
print("cetak bobot hasil learning")
print(W2)
print(W1)
A1 = 1 / (1 + np.exp (-(np.dot(data_test, W1))))
A2 = 1 / (1 + np.exp(-(np.dot(A1, W2))))
#Tebakan akhir setelah learning
print("Tebakan akhir setelah learning")
print(A2)
print(target_test)

for i in range(len(A2)): A2[i][0] = round(A2[i][0])
p = accuracy_score(target_test, A2)
print("Akurasi : ", p*100)
```
